```python
def cargos_mensuales_metodo_pago(df):
  # Agrupar por categor√≠a y calcular la facturaci√≥n total
  facturacion_por_categoria = df.groupby('PaymentMethod')['Charges.Monthly'].sum().sort_values(ascending=False)

  # Calcular porcentaje
  total_facturacion = facturacion_por_categoria.sum()
  percentual = facturacion_por_categoria.iloc[0] / total_facturacion * 100

  # Configura√ß√µes do gr√°fico
  plt.figure(figsize=(10, 6), facecolor= '#ffffff')
  ax = plt.axes()
  ax.set_facecolor('#ffffff')
  bars = plt.barh(facturacion_por_categoria.index, facturacion_por_categoria, color=Rojo_Intenso)

  # Agregar etiquetas y t√≠tulo
  plt.title('Cargos mensuales por metodo de pago', fontsize=18, color=P√∫rpura_Oscuro)
  plt.xlabel('')
  plt.ylabel('')
  plt.yticks(fontsize=12, color=Rosa_Intenso)
  plt.xticks([])
  ax.spines[['top', 'right', 'left', 'bottom']].set_visible(False)

  # Agregar total de facturaci√≥n al lado derecho de cada barra
  for bar in bars:
      plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
              f'MXN$ {bar.get_width()/1e6:.2f} M',
              ha='left', va='center', color=Rosa_Intenso, fontsize=12)

  # Agregar texto dentro del √°rea del gr√°fico
  plt.text(0.4, -1, f"$\\bf{percentual:.2f}$% de las ventas corresponden √∫nicamente a las ventas de Electronic check.",
          ha='left', va='center', color=P√∫rpura_Oscuro, fontsize=12)

  return plt
```

```python
cargos_mensuales_metodo_pago(df_account).show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

for i, gender in enumerate(['Female', 'Male']):
    data = df_customer[df_customer['gender'] == (0 if gender == 'Female' else 1)]
    counts = data['Partner'].value_counts(normalize=True)

    axs[i].pie(
        counts,
        labels=['No Partner', 'With Partner'],
        autopct='%1.1f%%',
        colors=[Rosa_Lavanda, Rojo_Intenso],
        startangle=90,
        explode=(0.05, 0),
        wedgeprops={'width': 0.4}
    )
    axs[i].axis('equal')
    axs[i].set_title(f'{gender} - Partner Proportion')

plt.suptitle('Distribuci√≥n de Clientes con y sin Pareja por G√©nero')
plt.show()
```

## üìå Introducci√≥n
Analizamos datos de clientes de Telecom X para identificar patrones asociados con la evasi√≥n (churn), con el fin de apoyar decisiones estrat√©gicas.

## üßπ Limpieza y Tratamiento de Datos

## üìä An√°lisis Exploratorio
- Proporci√≥n de evasi√≥n: X% de clientes abandonaron el servicio.
- Se observ√≥ mayor evasi√≥n entre clientes con contratos mensuales y m√©todos de pago electr√≥nicos.
- Factores como menor tiempo de contrato y mayor gasto mensual se correlacionan con mayor churn.

## üí° Conclusiones e Insights
- Los contratos de corto plazo y pagos autom√°ticos est√°n asociados a mayor churn.
- Los clientes con mayor gasto mensual tienen tasas de evasi√≥n elevadas, posiblemente por expectativas no satisfechas.
